```python
!pip install -q clearvoice gradio --upgrade
!apt -y -qq install ffmpeg

import gradio as gr
from clearvoice import ClearVoice

myClearVoice = ClearVoice(task='speech_enhancement', model_names=['MossFormer2_SE_48K'])

def process(path):
    output_wav = myClearVoice(input_path=path, online_write=False)
    out_path = "enhanced_output.wav"
    myClearVoice.write(output_wav, output_path=out_path)
    return out_path

demo = gr.Interface(
    fn=process,
    inputs=gr.Audio(type="filepath", label="Upload audio bị noise"),
    outputs=gr.Audio(label="Kết quả sau khi khử noise"),
    title="ClearerVoice-Studio – Khử noise & Nâng cấp giọng nói",
)
demo.launch(share=True)
```
